```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris

from kneed import KneeLocator
```

```python
df = pd.read_csv("online_shoppers_intention_num.csv")
df.head()
```

```python
# df = df.drop(['Revenue'],axis=1)
sns.set(rc={'figure.figsize':(20,10)})
sns.heatmap(df.corr().abs(),annot=True)
df = df.drop(['Administrative','Informational','ProductRelated','BounceRates','VisitorType=New_Visitor'],axis=1)
```

```python
df["Weekend"] = df["Weekend"].astype(int)

sc = StandardScaler()
df = pd.DataFrame(sc.fit_transform(df),columns = df.columns)
df.describe().round(2)
df.head()
```

Hierarchical

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


iris = load_iris()
X = iris.data

# setting distance_threshold=0 ensures we compute the full tree.
```

```python
link = 'ward'

model = AgglomerativeClustering(n_clusters=8, compute_distances=True, linkage=link)
model = model.fit(df)

plt.figure(figsize=(20,15))
plt.title("Hierarchical Clustering Dendrogram (Linkage: "+link+')')
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=4)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.savefig('img/hierarchical/'+link+'.png')
```

KMeans

```python
kmeans = KMeans(n_clusters=11, random_state=0).fit(df)
kmeans.labels_

kmeans.predict(df)

centers = kmeans.cluster_centers_
```

```python
df_centers = pd.DataFrame(centers,columns=df.columns)
df_centers.head()
df_centers = pd.DataFrame(sc.inverse_transform(df_centers),columns=df.columns)
df_centers = df_centers.round(1)
df_centers.to_csv('csv/kmeans/centers.csv')
```

Gaussian Mixture

```python
gm = GaussianMixture(n_components=4,covariance_type='diag',random_state=0)
labels = gm.fit_predict(df)
```

```python
def square(serie, variances):
    sq = np.square(variances)
    return np.sqrt(serie.multiply(sq))
```

```python
df_means = pd.DataFrame(gm.means_,columns=df.columns)
df_means = pd.DataFrame(sc.inverse_transform(df_means),columns=df.columns)
df_means = df_means.round(1)
df_means.to_csv('csv/gaussian/means.csv')
df_var = pd.DataFrame(gm.covariances_,columns=df.columns)
# df_var = df_var.apply(lambda x: square(x,sc.var_),axis=1)
df_var = df_var.round(2)
df_var.to_csv('csv/gaussian/variance_before.csv')
```

Elbow

```python
def elbow(sse):
    kl = KneeLocator(range(1, 11), 
                     sse, 
                     curve="convex", 
                     direction="decreasing")
    kl.plot_knee()
    return kl.elbow
```

```python
sse = []
for k in range(2, 11):
    gm = GaussianMixture(n_components=k, covariance_type='diag').fit(df)
    sse.append(gm.bic(df))

kl = KneeLocator(range(2, 11), 
                     sse, 
                     curve="convex", 
                     direction="decreasing")
kl.plot_knee()
print(kl.elbow)
```

```python
sse = []
for k in range(1, 15):
    km = KMeans(n_clusters=k).fit(df)
    sse.append(km.inertia_)
print(sse)
    
kl = KneeLocator(range(1, 15), 
                 sse, 
                 curve="convex", 
                 direction="decreasing")
kl.plot_knee()
print(kl.elbow)
```

Merging months

```python
def which_month(row):
    for col in df.columns:
        if 'Month=' in col and row[col] == 1:
            return col.split('Month=')[1]
    return 'Jan'
```

```python
df = pd.read_csv('clasif/gaussian.csv')
months = df.apply(lambda row: which_month(row), axis=1)
df['Month'] = months

for col in df.columns:
    if 'Month=' in col:
        df = df.drop([col],axis=1)

cluster = df['cluster'].copy()
df = df.drop(['cluster'],axis=1)
df['cluster'] = cluster
df.to_csv('clasif/gaussian_month.csv',index=False)
```
